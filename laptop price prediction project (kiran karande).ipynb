```python
def segment_processor(processor):
    if pd.isnull(processor):
        return 'Unknown'
    
    elif 'Intel Core i5 Processor (11th Gen)' in processor:
        return 'Intel i5 11 Gen '
    elif 'Intel Core i3 Processor (11th Gen)' in processor:
        return 'Intel i3 11 gen'
    elif 'Intel Core i5 Processor (10th Gen)' in processor:
        return 'Intel i5 10 gen'
    elif 'Intel Core i3 Processor (10th Gen)' in processor:
        return 'Intel i3 10 gen'
    elif 'Intel Core i5 Processor (12th Gen)' in processor:
        return 'Intel i5 12 gen'
    elif 'Intel Core i3 Processor (12th Gen)' in processor:
        return 'Intel i3 12 gen'
    elif 'Intel Core i7 Processor (12th Gen)' in processor:
        return 'Intel i7 12 gen'
    elif 'Intel Core i7 Processor (13th Gen)' in processor:
        return 'Intel i9 13 gen'
    elif 'Intel Core i9 Processor (12th Gen)' in processor:
        return 'Intel i9 12 gen'
    elif 'Apple M1 Processor' in processor or 'Apple M1 Pro Processor' in processor:
        return 'Apple M1'
    elif 'Apple M2 Processor' in processor or 'Apple M2 Processor' in processor:
        return 'Apple M2'
    elif 'AMD Ryzen 7 Hexa Core Processor' in processor:
        return 'AMD Ryzen 7'
    elif 'AMD Ryzen 9 Octa Core Processor (10th Gen)' in processor:
        return 'AMD Ryzen 9'
    elif 'AMD Ryzen 5 Hexa Core Processor (10th Gen' in processor:
        return 'AMD Ryzen 5'
    elif 'AMD Ryzen 3 Quad Core Processor' in processor:
        return 'AMD Ryzen 3'
    else:
        return 'Others'
    
    
newfeature['processor'] = newfeature['processor'].apply(segment_processor)
            
```

# Segmentation of the RAM

```python
newfeature['ram'].unique()
```

```python
def segment_ram(ram):
    if pd.isnull(ram):
        return 'Unknown'
    elif '4 GB DDR4 RAM' in ram or '4 GB LPDDR4X RAM' in ram or '4 GB LPDDR4 RAM' in ram:
        return '4 GB RAM'
    elif '8 GB DDR4 RAM' in ram or '8 GB LPDDR4X RAM' in ram or '8 GB Unified Memory RAM' in ram or '8 GB DDR5 RAM' in ram or '8 GB DDR3 RAM' in ram or '8 GB LPDDR3 RAM' in ram :
        return '8 GB RAM'
    elif '16 GB DDR4 RAM' in ram or '16 GB LPDDR5 RAM' in ram or '16 GB DDR5 RAM' in ram or '16 GB LPDDR4X RAM' or '16 GB Unified Memory RAM' in ram or '16 GB LPDDR4 RAM' in ram or '16 GB LPDDR3 RAM' in ram or '16 GB DDR3 RAM' in ram:
        return '16 GB RAM'
    elif '32 GB DDR5 RAM' in ram or '32 GB Unified Memory RAM' in ram or '32 GB DDR4 RAM' in ram or '32 GB LPDDR4X RAM' in ram:
        return '32 GB RAM'
    else:
        return 'Others'
    
    
newfeature['ram'] = newfeature['ram'].apply(segment_ram)
```

# Segmentation of the Operating system 

```python
newfeature ['os'].unique()
```

```python
def segment_os(os):
    if pd.isnull(os):
        return 'Unknown'
    elif 'Windows 11 Operating System' in os:
        return 'Windows 11'
    elif '64 bit Windows 11 Operating System' in os:
        return '64-bit Windows 11'
    elif 'Mac OS Operating System' in os:
        return 'Mac OS'
    elif 'Windows 10 Operating System' in os:
        return 'Windows 10'
    elif 'DOS Operating System' in os:
        return 'DOS OS'
    else:
        return 'Other'
    
    
newfeature['os'] = newfeature['os'].apply(segment_os)
```

# Segmentation of the STORAGE 

```python
newfeature['storage'].unique()
```

```python
def segment_storage(storage):
    if pd.isnull(storage):
        return 'Unknown'
    elif '512 GB SSD' in storage:
        return '512 GB SSD'
    elif '256 GB SSD' in storage:
        return '256 GB SSD'
    elif '4 TB SSD' in storage:
        return '4 TB SSD'
    elif '2 TB SSD' in storage:
        return '2 TB SSD'
    elif '1 TB SSD' in storage:
        return '1 TB SSD'
    elif '1 TB HDD|512 GB SSD' in storage:
        return '1 TB HDD|512 GB SSD'
    elif '2 TB HDD' in storage:
        return '2 TB HDD'
    else:
        return 'Other'
    
newfeature['storage'] = newfeature['storage'].apply(segment_storage)
```

# Model Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
linreg = LinearRegression()
knn = KNeighborsRegressor()
svm = SVR()
dt = DecisionTreeRegressor()
```

```python
from sklearn.metrics import r2_score
```

```python
def mymodel(model):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    train = model.score(X_train,y_train)
    test = model.score(X_test,y_test)
    
    print(f"Training score : {train}")
    print(f"Testing score  : {test}")
    
    print("Accuracy of model is :",r2_score(y_test,y_pred))
    
    return model
```
